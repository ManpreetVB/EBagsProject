# [dbo].[sp_AddToCart]

```sql
/****** Object:  StoredProcedure [dbo].[sp_AddToCart]    Script Date: 3/17/2024 11:12:00 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROC [dbo].[sp_AddToCart](@ID INT=null,@BagID INT,@BagBrand VARCHAR(100),@BagType VARCHAR(100),@Email VARCHAR(100)=null,
@Price DECIMAL(10,2),@Discount DECIMAL(10,2),@Quantity INT=null,@TotalPrice DECIMAL(10,2),@Output_Message VARCHAR(100) OUTPUT)
AS
 BEGIN
 DECLARE @UserID INT;  
 DECLARE @Price_ DECIMAL(10,2) 
  DECLARE @Discount_ DECIMAL(10,2);  
  DECLARE @TotalPrice_ DECIMAL(10,2) 
  

 SET @UserID = (SELECT 1 FROM Users WHERE Email = @Email);  
  SET @Price_ = (SELECT   1  FROM Bag WHERE ID = @ID);  
  SET @Discount_ = (SELECT (Price * @Quantity * Discount)/ 100 FROM Bag WHERE ID = @ID);  
  SET @TotalPrice_ = (SELECT (Price * @Quantity) - @Discount_ FROM Bag WHERE ID = @ID); 
  
   IF NOT EXISTS(SELECT 1 FROM Cart WHERE UserID = @UserID AND BagID = @ID) 
     BEGIN 
      INSERT INTO Cart(UserID,BagBrand,BagType,BagID,Price,Discount,Quantity,TotalPrice)
	  VALUES(@UserID,@BagBrand,@BagType,@BagID,@Price_,@Discount_,@Quantity,@TotalPrice_);

	  SET @Output_Message ='Item has been added to cart';
     END

   ELSE
        BEGIN
		UPDATE Cart SET Quantity = (Quantity + @Quantity) WHERE UserID = @UserID AND BagID = @ID;  
	   	 SET @Output_Message ='Item could not be added';
		 END
   END
   
   
GO
```

# [dbo].[sp_AddUpdateBag]

```sql
/****** Object:  StoredProcedure [dbo].[sp_AddUpdateBag]    Script Date: 3/17/2024 11:12:00 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROC [dbo].[sp_AddUpdateBag](@ID INT=null ,@BagBrand VARCHAR(100)=null,@Material VARCHAR(100)=null,@BagColor VARCHAR(100)=null,@Price INT=null,
@BagType VARCHAR(100)=null,@Discount DECIMAL(10,2)=null, @ImageUrl VARCHAR(100)=null,@Quantity INT=null,@IsActive INT=null,@ActionType VARCHAR(100)=null ,@Output_Message VARCHAR(100) OUTPUT )
AS 
BEGIN
 IF @ActionType ='ADD'
 BEGIN 
  INSERT INTO Bag(BagBrand,Material,BagColor,Price,BagType,Discount,ImageUrl,Quantity,IsActive)
  VALUES(@BagBrand,@Material,@BagColor,@Price,@BagType,@Discount,@ImageUrl,@Quantity,1);

  SET @Output_Message='Bag inserted successfully'
	
	 END
  ELSE
      BEGIN
	     SET @Output_Message='Bag not inserted'
	  END


IF @ActionType = 'UPDATE'
	BEGIN
		UPDATE Bag SET BagBrand=@BagBrand,Material=@Material,BagColor=@BagColor,Price=@Price,BagType=@BagType,Discount=@Discount,Quantity=@Quantity		
		WHERE ID = @ID;
		SET @Output_Message='Bag updated successfully'
	END
	ELSE
	BEGIN
	     SET @Output_Message='Bag not updated'
	  END

IF @ActionType = 'Delete'
	BEGIN
		UPDATE Bag SET IsActive = 0 WHERE ID = @ID;
		SET @Output_Message='Item deleted successfully'
	END

	ELSE
	BEGIN
	     SET @Output_Message='Item could not deleted'
	  END


IF @ActionType = 'Get'
	BEGIN
		SELECT * FROM Bag WHERE IsActive= 1;
		SET @Output_Message='Bag details fetched'
	END
     
	 ELSE
	BEGIN
	     SET @Output_Message='Bag details are not found'
	  END


IF @ActionType = 'GetByID'
	BEGIN
		SELECT * FROM Bag WHERE ID = @ID;
		SET @Output_Message='Item exists'
	END
	 ELSE
	BEGIN
	     SET @Output_Message='Item does not exist '
	  END

END
GO
```

# [dbo].[sp_CartList]

```sql
/****** Object:  StoredProcedure [dbo].[sp_CartList]    Script Date: 3/17/2024 11:12:00 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROC [dbo].[sp_CartList](@Email VARCHAR(100))
AS
BEGIN
    IF @Email != 'Admin'
	BEGIN
		SELECT C.ID, B.BagBrand, B.BagType, B.Price, B.Discount, C.Quantity, C.TotalPrice FROM Cart C 
		INNER JOIN Bag B ON B.ID = C.ID
		INNER JOIN Users U ON U.UserID = C.UserID
		WHERE U.Email =  @Email;
	END
	ELSE
	BEGIN
		SELECT B.ID, B.BagBrand, B.BagType, B.Price, B.Discount, B.Quantity,  0 AS TotalPrice FROM Bag B;
	END
END;
GO
```

# [dbo].[sp_OrderList]

```sql
/****** Object:  StoredProcedure [dbo].[sp_OrderList]    Script Date: 3/17/2024 11:12:00 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROC [dbo].[sp_OrderList](@UserID INT,
@Email VARCHAR(100) , @Type VARCHAR(100),@Output_Message VARCHAR(100) OUTPUT)
AS
BEGIN

 IF @Type = 'Admin'  
 BEGIN 
 DECLARE @ID INT =Null;
  SELECT O.ID,OrderNumber,OrderTotal,OrderStatus  
  
  ,CONCAT(U.FirstName,' ',U.LastName ) AS CustomerName 
  FROM Orders O INNER JOIN Users U  
  ON U.UserID = O.UserID;  
    SET @Output_Message='Order details fetched'
	END
     ELSE
      BEGIN
	     SET @Output_Message='Order details are not found'
	  END

	  IF @Type = 'User'  
 BEGIN  
  SELECT O.ID,OrderNumber,OrderTotal,OrderStatus  
  , CONCAT( U.FirstName,' ',U.LastName )  AS CustomerName 
  FROM Orders O INNER JOIN Users U  
  ON U.UserID = O.UserID  
  WHERE U.Email = @Email;
  SET @Output_Message='Order details fetched'
	END
     ELSE
      BEGIN
	     SET @Output_Message='Order details are not found'
	  END
  IF @Type = 'UserItems'  
 BEGIN  
  SELECT   
  O.ID, O.OrderNumber,O.OrderTotal,O.OrderStatus, B.BagBrand AS BagBrand, B.Material,B.BagColor,B.BagType,B.Price,OI.Quantity,OI.TotalPrice   
   ,CONCAT(U.FirstName,' ',U.LastName) AS CustomerName  
  , B.ImageUrl  
  FROM Orders O   
  INNER JOIN Users U ON U.UserID = O.UserID  
  INNER JOIN OrderItem OI ON OI.OrderID = O.ID  
  INNER JOIN Bag B ON B.ID = OI.ID  
  WHERE O.ID = @ID;  
    SET @Output_Message='Item details fetched'
	END
     ELSE
      BEGIN
	     SET @Output_Message='Item details are not found'
	  END   

END

GO
```

# [dbo].[sp_PlaceOrder]

```sql
/****** Object:  StoredProcedure [dbo].[sp_PlaceOrder]    Script Date: 3/17/2024 11:12:00 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
 CREATE PROC [dbo].[sp_PlaceOrder](@Email  VARCHAR(100))
   AS
   BEGIN
       DECLARE @OrderNumber VARCHAR(100);
	DECLARE @OrderID INT;
	DECLARE @OrderTotal INT;
	DECLARE @UserID INT;
	SET @OrderNumber =	(SELECT NEWID());
	SET @UserID = (SELECT UserID FROM Users WHERE Email = @Email);

	IF OBJECT_ID('tempdb..#TempOrder') IS NOT NULL DROP TABLE #TempOrder; 
	
	SELECT * INTO #TempOrder 
	FROM Cart WHERE UserId = @UserID;

	SET @OrderTotal = (SELECT SUM(TotalPrice) from #TempOrder);

	INSERT INTO Orders(UserID,OrderNumber,OrderTotal,OrderStatus)
	VALUES(@UserID,@OrderNumber,@OrderTotal,'Pending');

	SET @OrderID = (SELECT ID FROM Orders WHERE OrderNumber = @OrderNumber);

	INSERT INTO OrderItem(OrderID,BagBrand,Price,Discount,Quantity,TotalPrice)
	SELECT @OrderID, BagBrand,Price,Discount,Quantity,TotalPrice FROM #TempOrder;

	DELETE FROM Cart WHERE UserId = @UserID;
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[sp_Registration]    Script Date: 3/17/2024 11:12:00 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROC [dbo].[sp_Registration](@UserID INT,@FirstName VARCHAR(100),@LastName VARCHAR(100),@Email VARCHAR(100),@Password VARCHAR(100)
,@Address VARCHAR(1000),@Type VARCHAR(100),@IsActive INT,@Output_Message VARCHAR(100) OUTPUT)
As
BEGIN
	IF NOT EXISTS (SELECT 1 FROM Users WHERE  Email=@Email AND IsActive=1)
	BEGIN
		INSERT INTO Users(FirstName,LastName,Email,Password,Address,Type,IsActive)
		VALUES(@FirstName,@LastName,@Email,@Password,@Address,@Type,1);
		SET @Output_Message = 'User registered successfully';
	END
	ELSE
	BEGIN
		SET @Output_Message = 'User registration failed';
	END
	END
GO
```

# [dbo].[sp_RemoveToCart]

```sql
/****** Object:  StoredProcedure [dbo].[sp_RemoveToCart]    Script Date: 3/17/2024 11:12:00 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROC [dbo].[sp_RemoveToCart](@ID INT, @Email VARCHAR(100),@Output_Message VARCHAR(100) OUTPUT)    
AS    
BEGIN    
  DECLARE @UserID INT;    
   IF EXISTS (SELECT 1 FROM Users WHERE Email = @Email)
     BEGIN
      DELETE FROM Cart WHERE UserId = @UserId AND ID = @ID;
      SET @Output_Message ='Item has been removed from cart';
     END

    ELSE
     BEGIN
      SET @Output_Message ='Item could not be removed';
     END
END
GO
```

# [dbo].[sp_ViewUser]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ViewUser]    Script Date: 3/17/2024 11:12:00 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROC [dbo].[sp_ViewUser](@UserID INT=NULL,@Email VARCHAR(100)=NULL,@Output_Message VARCHAR(100) OUTPUT)
AS 
BEGIN
	IF @UserID IS NOT null AND @UserID != 0
	BEGIN
		SELECT * FROM Users WHERE UserID = @UserID AND IsActive= 1;
		SET @Output_Message='User exists'
	END
	 ELSE
      BEGIN
	     SET @Output_Message='User does not exist'
	  END

	IF @Email IS NOT null AND @Email != ''
	BEGIN
		SELECT * FROM Users WHERE Email = @Email AND IsActive = 1;
	SET @Output_Message='User exists'
	END
	 ELSE
      BEGIN
	     SET @Output_Message='User does not exist'
	  END
END
GO
```
